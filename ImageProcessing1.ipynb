```python
from PIL import Image       #Using the pillow library
```

```python
img = Image.open("wallpaper.jpg")       #Using Image to resize
img_resize = img.resize((200,200))
plt.imshow(img_resize)
```

```python
img_resize.save("/resized.jpg")
```

```python
import cv2     #Importing Opencv library
```

```python
img = cv2.imread("wallpaper.jpg")
print(img.shape)
```

```python
greyscale_img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
print(greyscale_img.shape)
```

```python
plt.imshow(greyscale_img, cmap='gray')
```

```python
cv2.imwrite("greyscale_wallpaper.jpg", greyscale_img)
```
